## FEATURE ENGINEERING

```python
## BATSMAN STRIKE RATE
batsman_stats = deliveries.groupby('batsman').agg({
    'batsman_runs': 'sum',
    'ball': 'count'
})
batsman_stats['strike_rate'] = (batsman_stats['batsman_runs'] / batsman_stats['ball']) * 100
batsman_stats.sort_values('strike_rate', ascending=False).head(10)
```

```python
# BOWLER STRIKE RATE
bowler_stats = deliveries.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'
})
bowler_stats['economy_rate'] = bowler_stats['total_runs'] / (bowler_stats['ball'] / 6)
bowler_stats.sort_values('economy_rate').head(10)
```

```python
## TOSS IMPACT-- UPGRATION VERSION OF FIRST UPGRADE
matches['toss_win_match_win'] = matches['toss_winner'] == matches['winner']

(matches['toss_win_match_win'].value_counts(normalize=True) * 100).plot(kind='bar')
plt.title("Toss Win vs Match Win (%)")
plt.ylabel("Percentage")
plt.savefig("./screenshot/percentage_of_toss_vs_match_win.png")
plt.show()
```

```python
#SEASON TREND ANALYSIS
matches.groupby('season')['id'].count().plot(kind='line')
plt.title("Matches per Season Trend")
plt.show()
#plt.savefig("./screenshot/Top5 players of the match award.png")
```

```python
#venue analysis
matches['venue'].value_counts().head(10).plot(kind='bar')
plt.title("Top Venues in IPL")
plt.show()
```

```python
## MULTIVARIABLE ANALYSIS
analysis = matches.groupby(['venue', 'toss_decision'])['winner'].value_counts()
analysis.head(10)
```

```python
# PREDICTIVE THINKING(RULE MODEL)
def predict_match(team1, team2):
    team1_wins = matches[matches['winner'] == team1].shape[0]
    team2_wins = matches[matches['winner'] == team2].shape[0]

    if team1_wins > team2_wins:
        return team1
    else:
        return team2
```

```python
# BUSINESS DECISION SIMULATION
toss_decision_analysis = matches.groupby('toss_decision')['winner'].value_counts(normalize=True) * 100
toss_decision_analysis
```
